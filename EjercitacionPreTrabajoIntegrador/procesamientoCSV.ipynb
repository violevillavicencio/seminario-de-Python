```
from google.colab import files
uploaded = files.upload()

from pathlib import Path
import csv

archivo_individuos = Path("usu_individual_T324.csv.txt")
archivo_hogares = Path("usu_hogar_T324.csv.txt")

# abro los archivos
# encoding="utf-8" para contemplar todos los caracteres y conservar su formato
# el delimitador del archivo es ;

with archivo_individuos.open(encoding="utf-8") as f:
    lector = csv.DictReader(f, delimiter=";")
    datos_individuos = list(lector)

with archivo_hogares.open(encoding="utf-8") as f:
    lector = csv.DictReader(f, delimiter=";")
    datos_hogares = list(lector)
# ------------------------------------------------------------------------------
# punto 1. informe cuántos varones y mujeres representa el dataset.
# campo CH04: Sexo, 1 representa a varon 2 a mujer
# pondera: cantidad de personas q representan lo mismo

total_varones = 0
total_mujeres = 0

for fila in datos_individuos:
    sexo = fila["CH04"]
    pondera = int(fila["PONDERA"])

    if sexo == "1":
        total_varones += pondera
    elif sexo == "2":
        total_mujeres += pondera

print("INFORME PUNTO 1:")
print(" total varones:", total_varones)
print(" total mujeres:", total_mujeres)

# ------------------------------------------------------------------------------
# punto 2. informe cuántas personas mayores de edad han completado los estudios
# secundarios.
# campo CH06: cuántos años cumplidos tiene
# campo NIVEL_ED: nivel de educación, 4 es secundario completo

total_mayores_secundario = 0

for fila in datos_individuos:
    edad = int(fila["CH06"])
    nivel = int(fila["NIVEL_ED"])
    pondera = int(fila["PONDERA"])

    if edad >= 18 and nivel == 4:
       total_mayores_secundario += pondera

print("INFORME PUNTO 2:")
print(" mayores de edad con secundario completo:", total_mayores_secundario)

# ------------------------------------------------------------------------------
# punto 3. informe el porcentaje de viviendas que son ocupadas por el/la
# propietario/a de la vivienda y del terreno.
# campo II7: Régimen de tenencia, 01 es Propietario de la vivienda y terreno

total_ponderado = 0
total_propietarias = 0

for fila in datos_hogares:
    tenencia = int(fila["II7"])
    pondera = int(fila["PONDERA"])
    total_ponderado += pondera

    if tenencia == 1:
        total_propietarias += pondera

porcentaje = (total_propietarias / total_ponderado) * 100

print("INFORME PUNTO 3:")
print(f" porcentaje de viviendas ocupadas por propietarios: {porcentaje:.2f}%")


# ------------------------------------------------------------------------------
# punto 4. informe el aglomerado con mayor y menor cantidad de viviendas con más
# de 2 ocupantes que no posean baño.
# campo IX_TOT, cantidad de miembros del hogar
# campo IV8, tiene baño 1 si 2 no
aglomerados = {}

for fila in datos_hogares:
    personas = int(fila["IX_TOT"])
    tiene_banio = int(fila["IV8"])
    aglo = fila["AGLOMERADO"]
    pondera = int(fila["PONDERA"])

    if personas > 2 and tiene_banio == 2:
        if aglo not in aglomerados:
            aglomerados[aglo] = 0
        aglomerados[aglo] += pondera

print("INFORME PUNTO 4:")

if aglomerados:
    aglo_mas = max(aglomerados, key=aglomerados.get)
    aglo_menos = min(aglomerados, key=aglomerados.get)

    print(" aglomerado con MÁS viviendas sin baño y con mas de 2 personas:", aglo_mas)
    print(" aglomerado con MENOS viviendas en esa condición:", aglo_menos)
else:
    print(" no se encontraron registros válidos para analizar este punto.")
```
